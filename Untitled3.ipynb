```python
import pandas as pd
from neuralprophet import NeuralProphet
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/seattle-weather.csv')
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
temp = df['temp_max']*df['temp_min']/2
temp_avg = pd.DataFrame(temp)
temp_avg.head()
```

```python
df.get('temp_avg')
```

```python
df.head()
```

```python
df['temp_avg'] = temp_avg
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df ['date'] = pd.to_datetime(df ['date'])
```

```python
plt.plot(df['date'],df ['temp_avg'])
plt.show()
```

```python
new_column = df[['date', 'temp_avg']] 
new_column.dropna(inplace=True)
new_column.columns = ['ds', 'y'] 
new_column.tail()
```

```python
n = NeuralProphet()
model = n.fit(new_column, freq='D')
```

```python
future = n.make_future_dataframe(new_column, periods=1500)
forecast = n.predict(future)
forecast.tail()
```

```python
plot = n.plot(forecast)
```
